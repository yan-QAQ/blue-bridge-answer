### 案例6-6 随机行走(ch4_9.py)
用程序模拟在格子里随意行走，从中心位置开始，然后在边缘处停下来。

假设格子大小是16*16，格子中两条线的距离为10个像素。程序首先绘制出灰色的格子，先绘制水平线，再绘制垂直线。
程序将笔移到中心位置，再使用while循环模拟随机行走。变量x和y用来跟踪其在格子中的当前位置。变量r生成一个0到3之间的随机数。每个数字代表一个方向。根据生成的方向，海龟向相应的方向移动，并且改变变量x和y的值。
当abs(x)或abs(y)为80时停止行走。


```python
import turtle
from random import randint

turtle.speed(1) # 设置海龟的速度为slowest

# 绘制16 * 16的网格
turtle.color("gray") # 网格的颜色
x = -80 
for y in range(-80, 80 + 1, 10):
    turtle.penup()
    turtle.goto(x, y) # 绘制水平线
    turtle.pendown()
    turtle.forward(160)

y = 80
turtle.right(90)
for x in range(-80, 80 + 1, 10):
    turtle.penup()
    turtle.goto(x, y) # 绘制垂直线
    turtle.pendown()
    turtle.forward(160)
    
turtle.pensize(3)
turtle.color("red")

turtle.penup()
turtle.goto(0, 0) # 回到中心
turtle.pendown()

x = y = 0 # 记录海龟的当前位置
while abs(x) < 80 and abs(y) < 80:    
    r = randint(0, 3)
    if r == 0:
        x += 10  # 向东走
        turtle.setheading(0)
        turtle.forward(10)      
    elif r == 1:
        y -= 10 # 向南走
        turtle.setheading(270)
        turtle.forward(10)      
    elif r == 2:
        x -= 10 # 向西走
        turtle.setheading(180)
        turtle.forward(10)      
    elif r == 3:
        y += 10 # 向北走
        turtle.setheading(90)
        turtle.forward(10)
```

### 案例6-7 猫捉老鼠(ch4_10.py)
该例子使用Turtle制作一个简单的小游戏，主要介绍Turtle的事件处理机制。

该游戏有两个角色，一个是猫，一个是老鼠，都是由小海龟表示，猫追逐老鼠，老鼠躲避，猫追上老鼠，游戏结束。猫和老鼠在一个方框里移动，不能超过方框的边界。猫由电脑自动控制，始终面向老鼠移动，老鼠由用户按上下左右键控制其移动，左键向左旋转，右键向右旋转，上键向前移动，下键向后移动。

Turtle库有两种使用方式，面向对象和面向过程。前面的案例我们都是使用的面向过程的函数实现，本例由于要使用两个小海龟，所以必须使用面向对象的方式实现。
mouse = turtle.Turtle()
cat = turtle.Turtle()

由于该案例是一个小游戏，需要与用户交互，因此我们要使用Turtle的事件处理机制。当用户按下键盘按键，点击鼠标时，系统都会产生相应的事件，通知上层程序做出相应的处理。事件处理机制，其实就是把事件和处理函数关联起来，当产生某种事件时，会自动调用相应的处理函数进行响应。
onkeypress(fun, key)，按键按下的事件处理关联函数，fun是处理函数，key是按键的字符串名字。

listen()函数，事件监听函数，调用此函数后，onkeypress()函数才会起作用。


```python
import time
import turtle

boxsize = 200
caught = False
score = 0
#老鼠前进
def up():
    mouse.forward(10)
    checkbound()
    
#老鼠左转
def left():
    mouse.left(45)
    
#老鼠右转
def right():
    mouse.right(45)
    
#老鼠后退
def back():
    mouse.backward(10)
    checkbound()
    
#结束游戏
def quitTurtle():
    window.bye()

#检测老鼠是否跑过界
def checkbound():
    global boxsize
    if mouse.xcor() > boxsize:
        mouse.goto(boxsize, mouse.ycor())  
    if mouse.xcor() < -boxsize:
        mouse.goto(-boxsize, mouse.ycor())
    if mouse.ycor() > boxsize:
        mouse.goto(mouse.xcor(), boxsize)
    if mouse.ycor() < -boxsize:
        mouse.goto(mouse.xcor(), -boxsize)

#
window = turtle.Screen()
mouse = turtle.Turtle()
cat = turtle.Turtle()

mouse.penup()                   #将老鼠的画笔提起，使不会出现轨迹
mouse.goto(100, 100)             #老鼠的初始地址放置在坐标（100，100）

window.onkeypress(up, "Up")      #按动方向键Up则执行函数up
window.onkeypress(left, "Left")
window.onkeypress(right, "Right")
window.onkeypress(back, "Down")
window.onkeypress(quitTurtle, "Escape")

#产生一个输入难度的对话框
difficulty = window.numinput("难度", "请输入游戏的难度（1~5）", minval=1, maxval=5)

paint = turtle.Turtle()
paint.penup()
paint.goto(-boxsize, boxsize)
paint.pendown()
for i in range(4):
    paint.forward(boxsize*2)
    paint.right(90)

paint.hideturtle()

#窗体监听按键
window.listen()

while not caught:
    cat.setheading(cat.towards(mouse))   #猫调整自己方向，使自己正对老鼠
    cat.forward(8 + difficulty)
    score += 1
    if cat.distance(mouse) < 5:            #老鼠与猫的距离少于5个像素就输了
        caught = True
    time.sleep(0.2 - (0.01 * difficulty))    #难度越高，运行速度越快
    
window.textinput("GAME OVER", "游戏得分:" + str(score * difficulty))
window.bye()
```
